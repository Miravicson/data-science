```python
import pandas as pd
```

```python
df = pd.read_excel('provider-list-08-2023.xlsx')
```

```python
df.head(3)
```

```python
column_map = {"Address":"address", "City":"city", "Retail":"retailTier", "Provider Name":"name", "S/N":"serialNumber", "SERVICE TYPE":"serviceType", "State":"state"}
df.rename(columns=column_map, inplace=True)
df['corporateTier'] = df['retailTier'] 
```

```python
def convert_to_title_case(string_or_other: str | int) -> str:
    if not isinstance(string_or_other, str):
        return string_or_other
    return string_or_other.title()


def apply_conversion_to_per_df_col(df: pd.DataFrame) -> None:
    for column in df.columns:
        df[column]= df[column].apply(convert_to_title_case)
        
```

```python
# convert column values to title case
apply_conversion_to_per_df_col(df)
df.head(3)
```

```python
len(df)
```

```python
df['city'].fillna(df['state'], inplace=True)
```

```python
len(df)
```

```python
filter_indices = df['city'].str.contains("/")
filter_indices = [i[0] for i in enumerate(filter_indices) if i[1]]
```

```python
df_with_multi_city = pd.DataFrame(df.loc[filter_indices, :].to_numpy(), columns=df.columns )
df_with_multi_city.tail()
```

```python
df.drop(filter_indices, inplace=True)
df.shape
```

```python

def explode_df_by_city(s: pd.Series, df: pd.DataFrame | None = None ) -> pd.DataFrame:
    df_cols = list(s.keys())
    combined_city = s.get('city')
    cities_arr = combined_city.split('/')
    new_df = pd.DataFrame(columns=df_cols) if df is None else df
    for city in cities_arr:
        new_dict = {**s, "city": city}
        new_df = new_df.append(new_dict, ignore_index=True)
        # new_df = pd.concat([new_df, new_dict])
        
    return new_df
        
        
    
```

```python
for i in df_with_multi_city.index:
    df = pd.concat([df, explode_df_by_city(df_with_multi_city.loc[i])])
```

```python
df.shape
```

```python
with open('provider-list.json', 'w') as provider_list:
    df.to_json(provider_list, orient="records")
    
```
